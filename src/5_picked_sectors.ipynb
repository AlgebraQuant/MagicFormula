```python
import pandas as pd
import matplotlib
```

```python
# Test file
df_magic = pd.read_csv('../data/magic_stocks.csv')
df_magic
```

```python
# Carregar tabela da CVM com os cadastros das empresas de capita aberto
CVM_PATH = "../data/external/cad_cia_aberta.csv"
df_sectors = pd.read_csv(
    CVM_PATH, sep=";",
    encoding="iso-8859-1",
    parse_dates=['DT_REG', 'DT_INI_SIT', 'DT_CANCEL']
)
cols = ['DENOM_SOCIAL', 'DT_REG', 'DT_CANCEL', 'SIT', 'DT_INI_SIT', 'CD_CVM',
       'SETOR_ATIV', 'CONTROLE_ACIONARIO']
df_sectors = (df_sectors
    .loc[:, cols]
    .sort_values('DT_INI_SIT', ignore_index=True)
    .drop_duplicates(subset="CD_CVM", keep="last")
    .sort_values("CD_CVM", ignore_index=True)
    .rename(columns={'CD_CVM': 'codcvm'})
    [['codcvm', 'SETOR_ATIV']]
)
df_sectors
```

```python
df_sectors.query('codcvm == 18724')
```

```python
df = pd.merge(df_magic, df_sectors)
df
```
